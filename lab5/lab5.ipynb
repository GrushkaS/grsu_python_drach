При решении научных задач с помощью численных методов зачастую приходится иметь дело с большими массивами данных в виде массивов и матриц. Стандартные средства Python (например, использование списков для хранения данных и прямой перебор элементов с помощью циклов) плохо подходят для таких задач, так как будут работать крайне медленно, в отличие от программ на компилируемых языках (Fortran, С). Пакет NumPy (http://www.numpy.org/) — это основной пакет, исполь¬зуемый в научных вычислениях в Python. NumPy предназначен для эф¬фективной работы с многомерными массивами, в том числе для эффек¬тивного применения математических операций к массивам и матрицам и их модификации без прямого перечисления элементов массивов. Постройте поточечный график функции. Сгенерируйте два массива для этого. Один со значениями по возрастанию (для X), второй со случайными значениями (для Y).



```python
import matplotlib.pyplot as pl

x = np.arange(0,10)
y = np.random.random(10)
print(pl.plot(x,y,':'))
```

Выберите тригонометрическую функцию $T(x)$, выберите полином $P(x)$. Постройте график функции $T(x)P(x)$.

```python
def P(x):
    return x^2
x = np.arange(0,10)
y = np.cos(x)*P(x)
print(pl.plot(x,y,':'))
```

К графику добавьте: обозначение оси абсцисс, обозначение оси ординат, название графика, вставка легенды (текста в label).

```python
pl.title('Graph')
pl.xlabel('x')
pl.ylabel('y')
pl.plot(x,y,'-',label = 'y = cos(x)*x^2')
pl.legend(loc='best')
```

Измените цвет графика на Ваш любимый. 

```python
pl.title('Graph')
pl.xlabel('x')
pl.ylabel('y')
pl.plot(x,y,'-',label = 'y = cos(x)*x^2', color = 'red')
pl.legend(loc='best')
```

Постройте несколько квивых. Первая - цвет красный, сплошная, вторая - цвет синий, штрих-пунктир.

```python
pl.plot(x, 2*x + 1, '-', color = 'red')
pl.plot(x, x**2, '-.', color = 'blue')
```

Постройте теже графики, только для первого добавьте маркеры из красных треугольников, для второго синие круглые маркеры.

```python
pl.plot(x, 2*x + 1, '-', color = 'red',marker = '^')
pl.plot(x, x**2, '-.', color = 'blue', marker = 'o')
```

Сделайте ее черно-белой

```python
black_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
pl.imshow(black_img)
```

```python
size_img = cv2.resize(img, (300, 300))
pl.imshow(size_img)
```

Разбейте ее на 2 картинки. Сохраните. 

```python
height, width = img.shape[:2]
up = img[:height//2, :width]
down = img[height//2:, :width]
cv2.imwrite('up.png', up)
cv2.imwrite('down.png', down)
```
